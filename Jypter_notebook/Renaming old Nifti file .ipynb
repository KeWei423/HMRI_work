```python
import os
import fnmatch
```

```python
wrkdir="/Volumes/Image_Repository/Research_Raw_NII/WH_Last_updated_30AUG19_KW/1.5T/"

num_folder = [folder for folder in os.listdir(wrkdir) if fnmatch.fnmatch(folder, '[0-9]*')]

for folder in num_folder:
    for path, subdirs, files in os.walk(wrkdir+folder):
        for direct in subdirs:
            if fnmatch.fnmatch(direct, '*FSPGR_3D*'):
                for file in os.listdir(path+'/'+direct):
                    if fnmatch.fnmatch(file, '[0-9]*[!A].nii.gz'):
#                         print(path+'/'+direct+'/'+file)
#                         print(path+'/'+direct+'/'+'WH'+folder+'_'+direct[direct.find('FSPGR'):]+'.nii.gz')
#                         os.rename(path+'/'+direct+'/'+file, 
#                                   path+'/'+direct+'/'+'WH'+folder+'_'+direct[direct.find('FSPGR'):]+'.nii.gz')
                        print("")
            if fnmatch.fnmatch(direct, '*T2_FLAIR*'):
                for file in os.listdir(path+'/'+direct):
                    if fnmatch.fnmatch(file, '[0-9]*[!A].nii.gz'):
                        print(path+'/'+direct+'/'+file)
                        print(path+'/'+direct+'/'+'WH'+folder+'_'+direct[direct.find('T2_FLAIR'):]+'.nii.gz')
                        os.rename(path+'/'+direct+'/'+file, 
                                  path+'/'+direct+'/'+'WH'+folder+'_'+direct[direct.find('T2_FLAIR'):]+'.nii.gz')
                        print("")
```
