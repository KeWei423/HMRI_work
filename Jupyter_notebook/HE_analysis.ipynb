```python
# import all files into 1 dictionary and conver that into a table
def create_database(path=os.getcwd()):
    subjects_dict={}
    for filename in os.listdir(path):
        if filename.endswith(".txt"):
            print("filename:", filename)
            ID=filename.split("_")[0]
#             print("ID: ",ID)
            f = open(os.path.join(path,filename))
            lines = f.read()
            if not len(lines)==0 :
                values = lines.replace("  \n","").split("  \t")
                numb_values = [float(numb) for numb in values] 
#                 print(numb_values)
                subjects_dict[ID]=numb_values
            else:
#                 subjects_dict[ID]=[]
                continue
        else:
            continue
    return subjects_dict
```

```python
def make_study_table(data_dict):
    title=["ID", "Cerebral_WM", "Cerebral_cortex", "Cerebellum_WM", "Cerebellum_cortex",
           "Caudate", "Putamen","globus_pallidus","Brain_stem", "Class"]
    study_table=Table(title)
    for key, values in data_dict.items():
        row=[key]
        row.extend(values)
        if key.startswith("BR"):
            row.append(0)
        else:
            row.append(1)
        study_table=study_table.with_row(row)
    return study_table
```

```python
def organize_files(path):
    
    header=["Index","SegId", "NVoxels_Volume_mm3", "StructName", "Mean", "StdDev", "Min", "Max", "Range"]
    counter=0
    db=pd.DataFrame()
    lable=[]
    for filename in os.listdir(path):
#         print("file: ", filenames)
        file=pd.read_csv(path+filename, comment="#", delim_whitespace=True, names=header)
        if counter==0:
            lable=file[['StructName']].T
            db=db.append(lable)
        row=file[['Mean']].T          
        row=row.rename({'Mean':filename[:-4]},axis='index')
        db=db.append(row)
        counter += 1
    db.names=lable
    return db
```
