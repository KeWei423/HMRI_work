```python
import os
from datascience import *
import numpy as np
import pandas as pd

import matplotlib.pyplot as plots
from mpl_toolkits.mplot3d import Axes3D
plots.style.use('fivethirtyeight')
%matplotlib inline
```

```python
def create_data(path):
    title=["ID","Left-Cerebral-White-Matter", "Left-Caudate", "Left-Putamen", "Left-Pallidum","Brain-Stem",
           "Right-Cerebral-White-Matter", "Right-Caudate", "Right-Putamen", "Right-Pallidum"]
    IDs=[]
    data=pd.DataFrame(columns=title)
    for file in os.listdir(path):
        if file.endswith(".txt"):
#             print(os.path.join(path, file))
            ID=file.split(".")[0]
#             print("ID: ", ID)
            IDs.append(ID)
            current=pd.read_csv(os.path.join(path, file),
                                sep='\t',
                                names=title,
                                skiprows=26, 
                                nrows=1, 
                                index_col=False)
#             print(current)
            data = data.append(current)
    data.ID = IDs
    data = data.set_index('ID')
    data=data.sort_values(by=['ID'])
    return data
```
